## Data Wrangling

This section is about inspecting our data and then cleaning them so that they can be used in our visualizations.

### General Properties

```python
mov_df = pd.read_csv('tmdb-movies.csv')
mov_df.head()
```

We have loaded the data and printed a few lines. It is obvious that our data is tidy. The revenue and budget have two columns each. The ones with the '_adj' ending, are re-calculated so that dollar inflation over time is accounted for. Those are the columns that we will use so that we have the same reference point. Based on the questions we have set we are interested in the columns below:

* *genres*
* *production_companies*
* *release_year*
* *vote_average*
* *budget_adj*
* *revenue_adj*


We can see that production_companies and genres contain several values seperated by "|", this is something that we will need to deal with in our analysis. We will also need to add an extra column with the calculated profits for each movies. First let's examine the data quality

```python
mov_df.info()
```

By printing this overview of our data it is evident that all data types are as desired. Also most of the features we are interested in are complete, except for genres and production_companies that have some missing values. Since those data types are descriptive, those missing values can not be imputed, and those rows will just be excluded from our analysis.

### Data Cleaning

Let's begin by adding the extra "profit" column that we will need. In order to easily substract bufget from revenue we will convert the two series to numpy arrays so that a vectorized substraction is performed.

```python
profit = mov_df.revenue_adj.to_numpy() - mov_df.budget_adj.to_numpy() # perform vectorized substraction
mov_df.insert(mov_df.shape[1], "profit_adj", profit, True) # add "profit_adj" as the last column
mov_df.head()
```

We can see that the column is added at the end of the dataframe as we intended. Next let's check our data for duplicates.

```python
print("Number of duplicates: {}".format(mov_df.duplicated().sum()))
```

We can see that one duplicate value is detected, let's print it and remove it

```python
dupl = mov_df[mov_df.duplicated()]
print(dupl)
mov_df.drop_duplicates(inplace=True)
print("\nDuplicate removed\n")
print("Number of duplicates {}".format(mov_df.duplicated().sum()))
print(mov_df[mov_df.original_title == "TEKKEN"])
```

It is the movie Tekken, after removing it we can see zero duplicates in our data, and we can still find this movie in the dataset, in the previous index. Finally let's handle the problem of the multiple values seperated by "|" in two of our columns. A good practice with dataframes in this case, is to split each of the rows in copies of itself that each contains a different value of this cell, as explained in <a href="https://stackoverflow.com/questions/29589977/pandas-how-to-convert-a-cell-with-multiple-values-to-multiple-row">this thread</a>. We'll repeat this process for both columns and create two extra dataframes, one for  each column, where the multivalued cells will be split to multiple rows as explained above. Instead of the method shown in the thread an easier fix in our case is to split the string in each cell and replace it with a list. The we can use the explode method to extend the dataframe.

```python
# let's begin with the genres
gen_list = mov_df.genres.str.split('|')         # split strings into lists
mov_df.genres = gen_list                        # replace strings with lists in the dataframe
mov_df.head()
```

We can see that lists are now elements of the 'genres' column. Let's explode those lists.

```python
mov_df_gen = mov_df.explode('genres') # explode based on the genres column
mov_df_gen.head()
```

After exploding we can see that the "genres" column is single-valued, our dataframe now has mure rows. Finally, let's do the same for the "production_companies"

```python
comp_list = mov_df.production_companies.str.split('|')         # split strings into list
mov_df.production_companies = comp_list                        # replace strings with lists in the dataframe
mov_df.head()
```

We can see that the "production_companies" column now contains lists.

```python
mov_df_comp = mov_df.explode('production_companies') # explode based on the production_companies column
mov_df_comp.head()
```

Now in this dataframe the "production_companies" column has single-valued cells.

<strong> Our cleaning process is now complete  and we have produced three different dataframes:
1. "mov_df_gen": A cleaned version of the original dataframe with the values of the "genres" column exploded in multiple rows. To be used for answering question 1
2. "mov_df": A cleaned version of the original dataframe, to be used for answering question 2
3. "mov_df_comp": A cleaned version of the original dataframe, where the values of the column "production_companies" are exploded into multiple rows. To be used for answering question 3 </strong>

## Exploratory Data Analysis


### Question 1: How do movie genres relate to ratings, popularity and profit?

To answer this question we need to group the "vote_average", "popularity" and "profit_adj" columns by the genres, with the average for each genre. We'll have to use the mov_df_gen dataframe as explained above.

```python
genre_ratings = mov_df_gen.groupby(['genres'])[["vote_average"]].mean() # group together and average
genre_profits = mov_df_gen.groupby(['genres'])[["profit_adj"]].mean()
genre_popularity = mov_df_gen.groupby(['genres'])[["popularity"]].mean()
genre_df = pd.concat([genre_ratings, genre_profits, genre_popularity], axis=1) # create new dataframe with variables of interest
genre_df
```

First let's visualize the sample size for each genre, this will help us gain an intuition of how reliable our results will be

```python
sn.countplot(y="genres", data=mov_df_gen, color='b');
```

It is evident that some genres like "Western", "TV Movie" and "Foreign" have very small sample size so we should be more cautious about results regarding those. Let's create some line plots to draw conclusions

```python
plt.stem(genre_df.index, genre_df.vote_average, use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("Ratings");
plt.xlabel("Genres");
plt.title("Average rating of different movie genres");
```

**Documentaries, animations, history, music and war films are generally highly rated, while horror movies are poorly rated**

```python
plt.stem(genre_df.index, genre_df.profit_adj, use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("profits in USD");
plt.xlabel("Genres");
plt.title("Average profits of different movie genres");
```

**Action, Adventure, Family, Fantasy and Science Fiction movies seem to be the most profitable. Some hihgly rated genres like documentaries, music and war movies show low profits**

```python
plt.stem(genre_df.index, genre_df.popularity, use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("popularity index");
plt.xlabel("Genres");
plt.title("Average popularity of different movie genres");
```

**Movie genres that showed high profits also tend to be popular and those with low profits tend to be unpopular.**

### Question 2: How do ratings, profit and popularity evolve in movies as years go by? Are there any trends?

Again we'll need to group the 3 columns by "release_year" column and average. Then we can normalize the three variables and use a multi-line plot to detect trends. We'll need to be cautious and use the mov_df dataframe, since we do not need genre or companies now, the extra, exploded rows would provide duplicate values in the variables that we are studying in this case.

```python
year_ratings = mov_df.groupby(['release_year'])[["vote_average"]].mean()
year_profits = mov_df.groupby(['release_year'])[["profit_adj"]].mean()
year_popularity = mov_df.groupby(['release_year'])[["popularity"]].mean()
year_df = pd.concat([year_ratings, year_profits, year_popularity], axis=1)
year_df.head()
```

```python
# let's normalize each column to z-scores
for name, column in year_df.items():
    year_df[name] = (year_df[name] - year_df[name].mean())/year_df[name].std(ddof=0)
    
year_df.head()
```

```python
plt.plot(year_df.index, year_df.vote_average, color='blue');
plt.plot(year_df.index, year_df.profit_adj, color='g');
plt.plot(year_df.index, year_df.popularity, color='r');
plt.title("Trends in movie characteristics over time");
plt.ylabel("Normalized units");
plt.xlabel("Years");
plt.legend(['ratings', 'profits', 'popularity']);
```

There are some clear trends:
* **Ratings and profits for movies seem to decrease over the years**
* **Popularity seems to increase over the years, in fact the last years there seems to be an exponential increase**

### Question 3: Are there any specific companies that have managed to be more profitable?

The "mov_df_comp" dataframe must be used here, we need to group "profit_adj" by "production_companies" and get it's sum to answer this question.

```python
comp_profits = mov_df_comp.groupby(['production_companies'])[["profit_adj"]].sum()
comp_profits
```

There are too many companies to display all of them in a plot, let's just trace the 10 most succesful. To do this we need to sort the data by profit.

```python
comp_profits.sort_values(by=['profit_adj'], ascending=False, inplace=True)
comp_profits
```

```python
plt.stem(comp_profits.index[0:9], comp_profits[0:9], use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("Average profit");
plt.xlabel("Commpany name");
plt.title("Average profits per movie for different production companies");
```

Some familiar names appear. **The profits of the most profitable companies is estimated to a few tens billions of dollars as we can see in the plot**

## Conclusions

In this analysis we studied data regarding thousands of movies. Some conclusions can be drawn from our visualizations:
1. Specific movie genres (Adventure, science fiction etc. ) seem to be more popular and more profitable, while some of the most highly rated genres (documentaries, music, history) seem to be less profitable and popular
2. While the popularity of movies in general seems to increase over the years, their profits and ratings seem to decrease. The profit decrease could be because of the increase in the movie budgets over the years.
3. The most famous production companies also show the largest profits (Warner Bros., Universal Pictures etc.), while their profits are estimated to be some tens of billions of dollars
