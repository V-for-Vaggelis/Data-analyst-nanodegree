## Preliminary Wrangling

The dataset used in this project, includes traffic data of 9 roads located in the city center of Thessaloniki from the years 2015 to 2020. These 9 particular roads, are located around the reference air-quality monitoring station which is the reason they are chosen.The data source is a project called TrafficThess for which you can find more information <a href="https://www.trafficthess.imet.gr/">here</a>. The data were accessed through the "historical data" option in this website. Originally the website provides one dataset for each road seperately, where the variables of interest are statistical parameters of the speeds of several taxis passing through the road and each observation differs by a 15 minute interval (or more) from the previous.The dataset provided here is an engineered version of all those datasets merged together. In detail, there were originally 9 datasets from which we used the variables mentioned below:
* timestamps : Time of the recording
* id : A number that is given by the TrafficThess project to each road
* road name : Name of the road in Greek characters
* average speed ($u$): The average speed of all the taxis within the time recording interval
* free flow speed ($u_f$): The speed at which a vehicle can move in the road at zero congestion conditions

In order to merge the 9 roads into 1 dataset we performed the following operations:
1. Upsampled the data for all the roads so that we have hourly observations
2. Estimated the traffic performance index (tpi) for each road, which scales from 0 (no traffic congestion) to 1 (full congestion) and is defined as $(u_f - u)/u_f$
3. Tagged each column with name of the road and the id in parentheses
4. "Downscaled" any tpi values that are greater than 1 to 1. In fact an outlier analysis was performed and it was determined that most outliers fall within the road's free flow speed limit, so removing outliers would be a loss of information. Also speed limits often get exceeded but that does not mean that those observations are errors.

Our goals through analyzing this dataset are:
1. To provide insights in the relationships between traffic conditions in different roads to answer as to how a single traffic congestion variable could be created to represent the whole area and serve as a variable to be compared with the air-pollution data provided by the station. For example, would averaging the 9 roads be a good choice? Or are there extreme differences between them.
2. to provide neat visualizations that communicate the traffic patterns in the city center of Thessaloniki. We can create extra categorical features that correspond to hour, day, and month to help capture and visualize patterns in the traffic data

## Data wrangling

> The data are pretty tidy, except from the timestamps each column represents the tpi of a road, and each row has the hourly average value. Some cleaniness issues are noted:
1. Greek characters and capital letters should be removed from the column names
2. The timestamps should be of datetime object type and not string
3. There are lots of missing values, we should examine their distribution and see if we can apply any imputation methods. One good option when working with variables of similar type is to use k-nn imputation.


First let's create a copy of the dataframe:

```python
tr_df_clean = traff_df.copy()
```

### Cleaniness issue 1
**Define**<br>
Replace capital letters and greek characters in the column names. In fact, the tpi column names aren't even descriptive regarding the value being represented, so we'll just use some new, descriptive names and map them to the old ones through a dictionary so that they can be used in the visualizations later on.<br><br>

**Code**

```python
tr_df_clean.rename(columns={"Timestamps": "timestamps"}, inplace=True)
tr_df_clean.set_index('timestamps', inplace=True) # temporaly set timestamps as index
old_cols = tr_df_clean.columns
tr_df_clean.columns = ["tpi_a", "tpi_b", "tpi_c", "tpi_d", "tpi_e", "tpi_f", "tpi_g", "tpi_h", "tpi_i"]
road_names = dict(zip(tr_df_clean.columns, old_cols))
tr_df_clean.reset_index(inplace=True) # set index back to original one
```

**Test**

```python
print(road_names)
tr_df_clean.head()
```

### Cleaniness issue 2
**Define**<br>
Change the data type of timestamps to datetime object<br><br>
**Code**

```python
tr_df_clean['timestamps'] = pd.to_datetime(tr_df_clean['timestamps'])
```

**Test**

```python
tr_df_clean.info()
```

### Adding extra features
> Since we are interested in exploring the traffic patterns in time, we can use the timestamps to extract several categorical variables that will help us produce interesting visualizations. Interesting insights could be provided by the variables below:
* hour of the day
* day of the week
* month of the year

Let's add those in our dataset

```python
tr_df_clean['hour'] = tr_df_clean.timestamps.dt.hour
tr_df_clean['day'] = tr_df_clean.timestamps.dt.strftime("%A") # extract day name
tr_df_clean['month'] = tr_df_clean.timestamps.dt.strftime("%B") # extract month name
tr_df_clean.head()
```

## Univariate Exploration

> In this section, we investigate distributions of individual variables. The 3 categorical variables drawn from the timestamps have a very predicatble behavior, so let's begin by dropping them for now

```python
roads_df = tr_df_clean.iloc[:,0:10]
roads_df.set_index('timestamps', inplace=True)
roads_df.head()
```

### Missing values visualization
We can use a heatmap to display the missing values patterns in our dataset. This can help us determine how to handle them from now on. <a href="https://dev.to/tomoyukiaota/visualizing-the-patterns-of-missing-value-occurrence-with-python-46dj">Here</a> is a good article with more information on how to visualize missing values.

```python
sb.heatmap(roads_df.isnull(), cbar=False);
plt.title("Missing values distribution");
```

The missing values appear as white areas in our heatmap. We can draw two conclusions:
* **The missing value patterns are continuous (no large chunks with missing values appear)**
* **There isn't any single variable with too many missing values (e.g more than 50% or so)**

That means that a method like k-nn can be used to impute the missing values. Doing so in this large dataset would require a lot of computaitonal power so we won't do it, bu it could be applied on smaller sub-sets of this data if needed. In the visualizations we'll create later on, we'll group by averaging on the categorical variables we have created, so the missing values are not expected to create any problems

### Distribution visualization
We can also plot the distribution of each one of the tpi variables to see if they behave similarily

```python
fig, axes = plt.subplots(nrows=3, ncols=3, sharex=True, sharey=True)
roads_df.hist(ax=axes);
fig.suptitle("Histograms for all roads", x=0.5, y=1.05, ha='center', fontsize='xx-large');
fig.text(0.5, 0.04, 'tpi', ha='center');
fig.text(0.04, 0.5, 'value counts', va='center', rotation='vertical');
```

* **All the distributions show similar behavior, they appear to be skewed to the left with a mode between 0.5-0.8**
* **One of the distributions seems to be bimodal with a second mode indicating a tendency for free flow, this probably has to do with the way we choose to deal with outliers in these data in the preliminary wrangling. Instead of deleting them, we scale them down to the free flow speed because we have determined that speed limits are sometimes exceeded and these scenarios indicate low congestion, so we wouldn't like to drop this kind of information.**

### Average daily profiles of roads
Traffic data are expected to show repeating patterns every day. Studying the average daily profiles of each road can be meaningful to determine if those patterns are the same for different roads. If so this would be an indication of good quality in the data and it could mean that different roads can be combined to produce a single variable.

```python
hourly_df = tr_df_clean.groupby('hour').mean()
# daily_df.drop('hour', axis=1, inplace=True)
filled_markers = ('o', 'v', '^', '<', '>', '8', 's', 'p', '*') # by default it can't handle more than 5 markers
sb.lineplot(data=hourly_df, dashes=False, markers=False);
plt.title("Average daily profile for all roads");
plt.ylabel("tpi");
```

* **Even though the roads display different congestion levels their profiles show similar behavior during the day**
* **This typical behavior is minimum congestions at approximately 02:00 - 06:00 and maximum at 12:00 - 19:00 which usually tends to have a small decrease between 15:00 - 18:00**

Those last observations are enough to indicate that a single tpi variable can be engineered to describe the traffic congestion of the area in general. __The average of all the available traffic nodes would be a good approximation, or perhaps a research  to determine if PCA analysis provides better results would be beneficial__. Those roads were manually picked so that they are approximately within a 100m radius of the air-pollution station. If automated access to all the traffic roads in the city is provided later on, a research on determining the ideal radius to pick and average data could be carried out by comparing the congestions estimated in different radii,  with the air-pollution data. At the moment, we are just interested on visualizing traffic patterns based on the available data. Let's provide a new dataframe with the average tpi and the engineered categorical variables.

```python
avg_tpi = roads_df.mean(axis=1)
avg_tpi.rename('avg_tpi', inplace=True)
tr_df_clean.set_index('timestamps', inplace=True)
avg_tpi_df = pd.concat([avg_tpi, tr_df_clean.hour, tr_df_clean.day, tr_df_clean.month], axis=1)
avg_tpi_df.head()
```

### Day and month average barplots
Interesting insights could occur by studying the average concentration of our feature of interest (avg_tpi) by day and month. Barplots could provide a good visualization for this case.

```python
daily_df = avg_tpi_df.groupby('day').mean().reset_index()
# daily_df.head()
first_color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0] # get the first default color of matplotlib
sb.barplot(x="day", y="avg_tpi", data=daily_df, color=first_color);
plt.title("Average tpi by day");
```

* **No large differences are observed in congestion by day. The largest difference is around 5%**
* **As expected in the weekends there is less traffic in the city, especially in Sunday**

```python
monthly_df = avg_tpi_df.groupby('month').mean().reset_index()
# daily_df.head()
first_color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0] # get the first default color of matplotlib
sb.barplot(x="month", y="avg_tpi", data=monthly_df, color=first_color);
plt.title("Average tpi by month");
```

* **No large differences are observed in congestion by month. The largest difference is around 5%**
* **As expected in the winter there is more traffic in the city and in August there is less**

## Multivariate Exploration

> We can create multivariate visualizations to see how smaller patterns in the tpi change during larger time periods. For example how would daily patterns change for different months. Heatmaps can be insightful in this case

### Daily tpi by month
We will create a heatmap to examine how the tpi scales for different days during different months

```python
multi_df1 = avg_tpi_df.groupby(['month', 'day']).mean().reset_index()
multi_df1.drop('hour', axis=1, inplace=True)
multi1 = multi_df1.pivot('month', 'day', 'avg_tpi') # scale to 2d table

print(multi1)
sb.heatmap(multi1);
plt.title("Average daily tpi by month");
```

* **Days and months that showed small congestions before, show the smallest congestions in total, when they occur at the same time (for example Sundays in summer months)**
* **Days and months that showed large congestions before, show the largest congestions in total, when they occur at the same time (for example fridays in December)**

### Hourly tpi by day
We will create a heatmap to examine how the tpi scales for different hours at different days

```python
multi_df2 = avg_tpi_df.groupby(['day', 'hour']).mean().reset_index()
multi2 = multi_df2.pivot('hour', 'day', 'avg_tpi')

print(multi2)
sb.heatmap(multi2);
plt.title("Average hourly tpi by day");
```

* **In general the findings agree with the ones in the average daily profiles for the single roads**
* **An interesting insight that was not clear before is that the maximum congestion occurs at Thursday and Tuesday afternoons.**

### Hourly tpi by month
We can also create a heatmap to visualize how the hourly congestions change by month

```python
multi_df3 = avg_tpi_df.groupby(['month', 'hour']).mean().reset_index()
multi3 = multi_df3.pivot('hour', 'month', 'avg_tpi')

print(multi3)
sb.heatmap(multi3);
plt.title("Average hourly tpi by month");
```

* **The previous foundings are confirmed once again since it is obvious that winter months in noon and afternoon hours are the most congested periods**

2. __The congestion levels at a road typically follow a left-skewed distribution.__

4. __Congestion levels usually show 2 peaks during the day: 12:00 - 03:00 and 18:00 - 20:00. Weekends and especially Sundays are typically connected with less congestion, and winter months (especially December) typically show the largest.__
