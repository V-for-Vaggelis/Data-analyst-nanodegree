## Investigation Overview

> This investigation aims at identifying patterns in the traffic congestion in the city center of Thessaloniki. Some questions are adressed:
1. What are the days and months of maximum and minimum congestion?
2. What are the hours of maximum and minimum congestion? 
3. How does the hourly beghavior change by day?



## Dataset Overview

> The dataset used in this project, includes traffic data of 9 roads located in the city center of Thessaloniki from the years 2015 to 2020. These 9 particular roads, are located around the reference air-quality monitoring station which is the reason they are chosen. The data source is a project called TrafficThess for which you can find more information <a href="https://www.trafficthess.imet.gr/">here</a>. The data were accessed through the "historical data" option in this website. Originally the website provides one dataset for each road seperately, where the variables of interest are statistical parameters of the speeds of several taxis crossing the road and each observation differs by a 15 minute interval (or more) from the previous.The dataset provided here is an engineered version of all those datasets merged together. The average speed for each observation and the free flow speed for each road are accessed so that we can estimate a congestion index called tpi (traffic performance index). The data are also upsampled so that we have hourly observations. So we have as a result a dataset indexed by hourly timestamps and with 9 columns holding the congestion index for each road.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# setting plot styles
plt.style.use('seaborn-whitegrid')
plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams["font.size"] = 22

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
traff_df = pd.read_csv("traffic_thess_data.csv")
tr_df_clean = traff_df.copy()

# some cleaning issues
tr_df_clean.rename(columns={"Timestamps": "timestamps"}, inplace=True)
tr_df_clean.set_index('timestamps', inplace=True) # temporaly set timestamps as index
old_cols = tr_df_clean.columns
tr_df_clean.columns = ["tpi_a", "tpi_b", "tpi_c", "tpi_d", "tpi_e", "tpi_f", "tpi_g", "tpi_h", "tpi_i"]
road_names = dict(zip(tr_df_clean.columns, old_cols))
tr_df_clean.reset_index(inplace=True) # set index back to original one

tr_df_clean['timestamps'] = pd.to_datetime(tr_df_clean['timestamps'])

# adding the extra features
tr_df_clean['hour'] = tr_df_clean.timestamps.dt.hour
tr_df_clean['day'] = tr_df_clean.timestamps.dt.strftime("%A") # extract day name
tr_df_clean['month'] = tr_df_clean.timestamps.dt.strftime("%B") # extract month name

# getting the average of all roads
roads_df = tr_df_clean.iloc[:,0:10]
roads_df.set_index('timestamps', inplace=True)
avg_tpi = roads_df.mean(axis=1)
avg_tpi.rename('avg_tpi', inplace=True)
tr_df_clean.set_index('timestamps', inplace=True)
avg_tpi_df = pd.concat([avg_tpi, tr_df_clean.hour, tr_df_clean.day, tr_df_clean.month], axis=1)
```

## Daily average barplot

> The average congestion for all days scales between 0.5 and 0.6 and the differnces between them are small. Weekends show smaller congestions and specifically Sunday seems to be the less congested day.

```python
daily_df = avg_tpi_df.groupby('day').mean().reset_index()

# re-order based on days: source: https://stackoverflow.com/questions/35193808/re-order-pandas-series-on-weekday
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mapping = {day: i for i, day in enumerate(days)}
key = daily_df['day'].map(mapping)
daily_df = daily_df.iloc[key.argsort()]

first_color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0] # get the first default color of matplotlib
sb.barplot(x="day", y="avg_tpi", data=daily_df, color=first_color);
plt.title("Average tpi by day");
plt.ylim(top=1);
plt.xticks(rotation=90);
plt.ylabel("average tpi");
```

## Monthly average barplot

> No large differences are observed between months as well. Autumn and winter months (especially December) show larger congestions while summer months (especially August) show smaller congestions

```python
monthly_df = avg_tpi_df.groupby('month').mean().reset_index()

months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
mapping = {month: i for i, month in enumerate(months)}
key = monthly_df['month'].map(mapping)
monthly_df = monthly_df.iloc[key.argsort()]

first_color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0] # get the first default color of matplotlib
sb.barplot(x="month", y="avg_tpi", data=monthly_df, color=first_color);
plt.title("Average tpi by month");
plt.ylim(top=1);
plt.xticks(rotation=90);
plt.ylabel("average tpi");
```

## Colormap of hourly average by day
> It is clear that Tuesday and Thursday afternoons (around 19:00) are the most congested periods. Fridays seem to have the largest congestions at noon. All days seem to typically have one peak at noon (around 13:00) and one at afternoon (around 19:00). Finally around 05:00 is when the least congestion occurs.


```python
multi_df2 = avg_tpi_df.groupby(['day', 'hour']).mean().reset_index()


# print(multi_df2.tail())
multi2_df = multi_df2.pivot('hour', 'day', 'avg_tpi')
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']


multi2_df = multi2_df[days]
sb.heatmap(multi2_df);
plt.title("Average tpi by hour by day");
plt.xticks(rotation=90);
locs, labels = plt.yticks() # get previous ticks location
plt.yticks(locs, ('00:00', '02:00', '04:00', '06:00', '08:00', '10:00', '12:00', '14:00', '16:00', '18:00', '20:00', '22:00'),
          rotation=0);
```
