# Project 4: Merging and wrangling data from different sources

## Introduction
This project is about combining and cleaning different data sources to make them easily usable for visualization, models etc. The data are about tweets of dogs, specifically there are three different data sources:
* The WeRateDogs Twitter archive, provided as a csv file called **twitter-archive-enhanced.csv**
* The tweet image predictions, i.e., what breed of dog (or other object, animal, etc.) is present in each tweet according to a neural network. These data are hosted in the Udacity servers and accessed through an http request. The response is a tsv file called **image_predictions.tsv**
* Each tweet's retweet count and favorite ("like") count, accessed through tweeter's API and stored programmatically in a txt file called **tweet_json.txt**.

The project aims goals are set below:
1. Create and access the above data sources properly
2. Identify and fix some cleaniness and tidyness issues in them
3. Merge all the data together and provide a master csv file called **twitter_archive_master.csv**.
3. Provide some insights and visualizations to demonstrate the usability of these data

##### Tidyness issue 1
**Description**<br>
Keep only breed predictions that correspond to a dog breed and have a conf interval > 0.5<br><br>
**Code**

```python
breed_pred_list = []
breed_conf_list = []
breed_list = []
for i, row in pred_df.iterrows():
    
    dog_breeds = np.array([row[3], row[6], row[9]])
    dog_conf = np.array([row[4], row[7], row[10]])
    dog_bool = np.array([row[5], row[8], row[11]])
    
    dog_breeds = dog_breeds[dog_bool] # keep only predictions that correspond to a dog breed
    dog_conf = dog_conf[dog_bool]
    
    
    dog_breeds = dog_breeds[np.where(dog_conf > 0.5)] # keep only conf > 0.5
    dog_conf = dog_conf[np.where(dog_conf > 0.5)]
    if dog_conf.shape[0] > 0: # if there are any predictions left push array to list
        breed_conf_list.append(dog_conf)
        breed_pred_list.append(dog_breeds)
    else: # if no pred left create and push nan array
        breed_conf_list.append(np.array([np.nan]))
        breed_pred_list.append(np.array([np.nan]))

# breed_pred = pd.Categorical(breed_pred_list, categories=breed_list, ordered = False) # this should fill NaN's properly
pred_clean_df['breed_pred'] = breed_pred_list
pred_clean_df['breed_conf'] = breed_conf_list
pred_clean_df.head()
```

Since this algorithm provides low confidence intervals, it is very possible that this filtering we applied only outputs 1 or less valid predicted breeds for each tweet, in this case we could easily flatten our cells so that they are not represented by numpy arrays. Let's check if that applies

```python
multivalued = pred_clean_df.breed_pred.apply(lambda x: x.shape[0] > 1)
print("Number of cells with 2 or more predicted breeds: {}".format(multivalued.sum()))
```

So we can flatten every cell in the "breed_pred" and "breed_conf" columns

```python
pred_clean_df.breed_pred = pred_clean_df.breed_pred.apply(lambda x: x[0])
pred_clean_df.breed_conf = pred_clean_df.breed_conf.apply(lambda x: x[0])
```

Finally let's delete the old columns related to breed predictions

```python
pred_clean_df.drop(['p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog'], axis=1, inplace=True)
```

###### 1c: Gathering data from twitter api

Let's begin by creating an api object. The user keys and passwords are imported from a local script so they are not visible, so this code will only work in a local machine that can provide this script, to test it enter you keys and passwords manually. Let's request the first tweet to see what kind of response we'll get:

```python
from my_twitter_codes import get_twitter_codes # imported locally
import tweepy
import json


consumer_key, consumer_secret, access_token, access_secret = get_twitter_codes() # to run it comment this line, uncomment the line below and fill it with your access info
# consumer_key, consumer_secret, access_token, access_secret = 

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

try:
    tweet = api.get_status(arch_clean_df.tweet_id[10], tweet_mode='extended')
    print(tweet._json)
except:
    print("Tweet not found")
```

Great, now we need to do this proccess iteratively and write each json response in a line of a txt file called **tweet_json.txt**. if a tweet is not found it means it's been deleted, we'll save those id's just in case but this will be handled automatically when merging with the previous data. Also we might have to deal with request-rate limits in the twitter api. The wait_on_rate parameters we set above might help with that but some tuning might still ne neccessary. <a href="https://stackabuse.com/reading-and-writing-json-to-a-file-in-python/">Here</a> is a good resource on how to store and access json data. Let's try it out, the expected number of fetched tweets is:

```python
print("Expected tweets: {}".format(arch_clean_df.shape[0]))
```

```python
if not os.path.exists('tweet_json.txt'):

    count = 0 # count the found tweets
    deleted_tweets = []
    for i, row in arch_clean_df.iterrows():
        
        try: # if tweet is found save it a txt line and print it's id
            tweet = api.get_status(row[0], tweet_mode='extended')
            with open('tweet_json.txt', 'a') as file:
                json.dump(tweet._json, file)
                file.write('\n')
            print(tweet._json['id'])
            count += 1
        except: # if tweet is not found store it's id and print a message
            deleted_tweets.append(row[0])
            print("Tweet not found")
            
    print("File created, tweets found: {}".format(count))
else:
    print("File already exists")
```

We see all the printed id's, and the message for the ones not found. Most of them were found. Also each time we reached the request limit the code would alert and pause and wait, thanks to the parameters we set in the api object. The desired file is now located in our working directory, let's try and access it, for simplicity reasons we'll only store the three columns we're interested in, that is "tweet_id", "retweet_count", "favorite_count". Let's also rename the  "favorite_count" column to "fav_count" since the name is too large.

```python
data = {'tweet_id': [], 'retweet_count': [], 'fav_count': []} # save all the data we need in a dictionary

with open('tweet_json.txt') as json_file:

    for line in json_file: # each line is a json file holding a tweet
        tweet = json.loads(line)
        data['tweet_id'].append(tweet['id'])
        data['retweet_count'].append(tweet['retweet_count'])
        data['fav_count'].append(tweet['favorite_count'])

tweet_api_df = pd.DataFrame.from_dict(data) # create df from dict
tweet_api_df.head()
```

### Part 3: Visualizations

##### 3a: Visualization of the 10 most favorite dog breeds

To answer this question we need to group the "fav_count" column by the "breed_pred", with the average for each breed. Then we'll need to sort based on the "fav_count"

```python
breed_fav = twitter_archive_master_df.groupby(['breed_pred'])[["fav_count"]].mean() # group together and average
breed_fav.sort_values(by=['fav_count'], ascending=False, inplace=True)
breed_fav.head()
```

```python
plt.stem(breed_fav.index[0:9], breed_fav[0:9], use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("Average favorite tags");
plt.xlabel("Dog breed");
plt.title("Average number of likes for different dog breeds");
```

##### 3b: Visualization of the different dog-state favourite counts

To answer this question we need to group the "fav_count" column by the "dog_state", with the average for each state. Then we'll need to sort based on the "fav_count".

```python
state_fav = twitter_archive_master_df.groupby(['dog_state'])[["fav_count"]].mean() # group together and average
state_fav.sort_values(by=['fav_count'], ascending=False, inplace=True)
state_fav.head()
```

```python
plt.stem(state_fav.index, state_fav, use_line_collection=True);
plt.xticks(rotation=90);
plt.ylabel("Average favorite tags");
plt.xlabel("Dog state");
plt.title("Average number of likes for different dog states");
```
